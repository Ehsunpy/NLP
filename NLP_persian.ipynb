- 🧩 متن ورودی به کلمات توکن‌سازی می‌شود و هر کلمه که در لیست توقف‌ها و کاراکترهای نقطه‌گذاری نیست، به لیست توکن‌ها اضافه می‌شود.
- 🔄 لیست توکن‌های نهایی برگردانده می‌شود.

📊 **اجرا و نمایش توکن‌های پردازش شده:**

- لیست توکن‌های پردازش شده توسط تابع `process_text` برای متغیر `text` و نمایش نتیجه.

این کد به شما کمک می‌کند تا متون فارسی خود را پردازش کنید و توکن‌های مفید را استخراج کنید.


````python
# prompt: def process_text(text):
#   tokenize = []
#   text = text.lower()
#   for word in nltk.word_tokenize(text):
#     if word not in stopword and word not in string.punctuation:
#       tokenize.append(word)
#   return tokenize
# این با کتابخانه hazm

from hazm import stopwords_list, word_tokenize
import string
import re

# Assuming hazm is already installed (pip install hazm)

persian_stopwords = stopwords_list()

def process_text(text):
    tokenize = []
    text = text.lower()
    for word in word_tokenize(text):
        if word not in persian_stopwords and word not in string.punctuation:
            tokenize.append(word)
    return tokenize


tokens = process_text(text)
tokens
````

🔄 این تابع متن ورودی را به حروف کوچک تبدیل کرده، سپس کلمات را توکن‌سازی می‌کند و هر کلمه‌ای که در لیست توقف‌ها و کاراکترهای نقطه‌گذاری نیست، به لیست توکن‌ها اضافه می‌شود.
📝 نهایتا، لیست توکن‌های پردازش شده را برمی‌گرداند و این لیست شامل کلمات مفید متن خواهد بود.

````python
# prompt: این که میدم میخوام معادل بشه با کتابخانه hazmscore_sentences = {}
# for sent in text:
#   for word in nltk.word_tokenize(sent.lower()):
#     if word in word_freq.keys():
#       if sent not in score_sentences.keys():
#         score_sentences[sent] = word_freq[word]
#       else:
#         score_sentences[sent] += word_freq[word]
# score_sentences

hazm_score_sentences = {}
for sent in text:
  for word in hazm.word_tokenize(sent.lower()):
    if word in freq_dist.keys():
      if sent not in hazm_score_sentences.keys():
        hazm_score_sentences[sent] = freq_dist[word]
      else:
        hazm_score_sentences[sent] += freq_dist[word]
hazm_score_sentences
````

📥 کتابخانه `heapq` را با استفاده از `import heapq` ایمپورت می‌کنیم.

🔝 با استفاده از `heapq.nlargest(5, hazm_score_sentences, key=hazm_score_sentences.get)`، پنج جمله با بالاترین امتیاز را از دیکشنری `hazm_score_sentences` استخراج می‌کنیم و نتیجه را در متغیر `best_sentences` ذخیره می‌کنیم.
